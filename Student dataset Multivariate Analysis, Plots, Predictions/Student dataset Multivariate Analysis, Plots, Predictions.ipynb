## Splitting the data into Test and Train sets

```python
from collections import Counter
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=44)
```

## Oversampling our dataset

```python
from imblearn.combine import SMOTETomek

from imblearn.over_sampling import RandomOverSampler

smt = SMOTETomek(random_state=92)
ros = RandomOverSampler(random_state=12)

def oversampler(X,y):
    X_rs, y_rs = smt.fit_resample(X,y)
    print('Before OS: ', Counter(y), ' \nAfter OS: ', Counter(y_rs))
    return X_rs, y_rs
```

```python
X_train_rs, y_train_rs = oversampler(X_train,y_train)
```

## Now we will use a Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier

def rf_train_predict(X_train,y_train,X_test, y_test, rstate, os):
    reg = RandomForestClassifier(max_depth=3,random_state=rstate)
    if os == 1:     ### If os (oversampling) is set to 1, we oversample the train data
        X_train, y_train = oversampler(X_train, y_train)   
        reg.fit(X_train, y_train)
    else:
        reg.fit(X_train, y_train)
    return reg, reg.predict(X_test), reg.score(X_test, y_test)
```

I created the oversample option to compare the performance with and without oversampling

```python
rf_reg, rf_y_pred, rf_score = rf_train_predict(X_train, y_train, X_test, y_test, 12, os=1) ### Here we have the regressor, the predictions, and the accuracy score
```

```python
def regressor_fe_score(regressor,top_nr):
    reg_feature_importance = pd.concat([pd.DataFrame(X.columns), pd.DataFrame(regressor.feature_importances_)],axis=1)
    reg_feature_importance.columns = ['Feature','Score']
    reg_feature_importance.sort_values(by=['Score'],ascending=False)
    reg_top_features = reg_feature_importance.nlargest(top_nr,'Score')
    return reg_top_features
```

```python
reg_top_features = regressor_fe_score(rf_reg,10)
reg_top_features
```

### Cross Validation -  Stratified KFold

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
skfold = StratifiedKFold(n_splits=4)
print('F1 Score: ', cross_val_score(rf_reg, X_test,y_test, cv=skfold, n_jobs=-1,scoring='f1_macro').mean(), ', dev:',cross_val_score(rf_reg, X_test,y_test, cv=skfold, n_jobs=-1,scoring='f1_macro').std())
print('Accuracy: ', cross_val_score(rf_reg, X_test,y_test, cv=skfold, n_jobs=-1,scoring='accuracy').mean(), ', dev:',cross_val_score(rf_reg, X_test,y_test, cv=skfold, n_jobs=-1,scoring='accuracy').std())
```

### Performance Metrics

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def performance_metrics(y_test,y_pred):
    print('Total errors: ', (y_test!=y_pred).sum())
    print(confusion_matrix(y_test,y_pred))
    print(accuracy_score(y_test,y_pred))
    print(classification_report(y_test,y_pred))
```

## Let's try Logistic Regression

NOTE! Regarding the fact that Logistic Regression uses Gradient Descent, we will have to scale down our data.

```python
# Sdandard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

def log_train_predict(X_train,y_train,X_test, y_test, os):
    reg = LogisticRegression()
    scaler = StandardScaler()
    X_test = scaler.fit_transform(X_test)

    if os == 1:     ### If os (oversampling) is set to 1, we oversample the train data
        X_train, y_train = oversampler(X_train, y_train)
        X_train = scaler.fit_transform(X_train)
        reg.fit(X_train, y_train)
    else:
        X_train = scaler.fit_transform(X_train)
        reg.fit(X_train, y_train)
    return reg, reg.predict(X_test), reg.score(X_test, y_test)
```

```python
#Predict
log_reg, log_y_pred, log_score = log_train_predict(X_train,y_train,X_test, y_test, os=1)
log_score
```

## Another Approach
Due to the fact that we have in imbalanced dataset, meaning that few students don't pass the class, let's consider that the studentds that don't pass as outliers and try to detect them.

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from collections import Counter

Counter(y)
```

```python
### Invert the labels: 1 = is outlier
map = {1:0, 0:1}
y.map(map)
Counter(y)
```

```python
contamination = Counter(y)[0] / Counter(y)[1]
contamination
```

```python
forest = IsolationForest(n_estimators=100, contamination=contamination, random_state=99)
forest.fit(X.to_numpy(),y.to_numpy())
y_pred_forest = forest.predict(X.to_numpy())
Counter(y_pred_forest)
```

```python
def convert(y_pred):
    y_pred[y_pred==1] = 0
    y_pred[y_pred==-1] = 1
    print(Counter(y_pred))
    return y_pred
```

```python
y_pred_forest = convert(y_pred_forest)
```

```python
performance_metrics(y,y_pred_forest)
```
