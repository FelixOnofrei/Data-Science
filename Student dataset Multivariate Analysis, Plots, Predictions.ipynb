## Feature Selection
### Determining feeature importances

1. Univariate Selection - Using CHi2


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

### Function to select the KBest features using chi2 function
def top_feature_func(X,y,top_nr):                       
    best_features = SelectKBest(score_func=chi2, k=10)
    fit = best_features.fit(X, y)
    df_scores = pd.DataFrame(fit.scores_)
    df_columns = pd.DataFrame(X.columns)
    scores = pd.concat([df_columns, df_scores],axis=1)
    scores.columns = ['Feature','Score']
    scores.sort_values(by=['Score'],ascending=False)
    top_features = scores.nlargest(top_nr,'Score')
    return top_features
```

## Oversampling our dataset

```python
from imblearn.combine import SMOTETomek
from collections import Counter

smt = SMOTETomek(random_state=9)

X_res, y_res = smt.fit_resample(X,y)
```

## Splitting the data into Test and Train sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=2)
```

## Now we will use a Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
reg = RandomForestClassifier(max_depth=3,random_state=2)
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
y_pred
```

```python

reg_feature_importance = pd.concat([pd.DataFrame(X_res.columns), pd.DataFrame(reg.feature_importances_)],axis=1)
reg_feature_importance.columns = ['Feature','Score']
reg_feature_importance.sort_values(by=['Score'],ascending=False)
reg_top_features = reg_feature_importance.nlargest(15,'Score')
```
